```python
import pandas as pd
import csv
```

```python
airlines_path = "Dataset/airlines.csv"
airlines_df = pd.read_csv(airlines_path)

airports_path = "Dataset/airports.csv"
airports_df = pd.read_csv(airports_path)

flights_path = "Dataset/flights.csv"
```

```python
#Flights data set contained rows where the origin airport and destination airport code were integers. 
#Dropping those rows from while reading data into the dataframe.

list_int =[]
with open(flights_path) as flights_fo:
    flights_csv = csv.reader(flights_fo)
    next(flights_csv)
    index= 1
    for row in flights_csv:
        try:
            int_val = int(row[7])
            list_int.append(index)
        except:
            pass
        index = index+1
```

```python
flights_df = pd.read_csv(flights_path,skiprows=list_int)
flights_df.info()
```

```python
#Since few columns contain multiple null values dropping the columns

modflights_df = flights_df.drop(columns=['CANCELLATION_REASON','AIR_SYSTEM_DELAY','SECURITY_DELAY',
                                         'AIRLINE_DELAY','LATE_AIRCRAFT_DELAY','WEATHER_DELAY'])
modflights_df.info()
```

```python
#Creating primary key "flight id" in the modflights dataset with the index column

modflights_df['FLIGHT_ID'] = modflights_df.index
modflights_df.set_index('FLIGHT_ID', inplace=True)
modflights_df.head()
```

```python
#Matching the column names in the modflights dataset with the airports dataset and airlines dataset
modflights_df = modflights_df.rename(columns={"AIRLINE": "AIRLINE_CODE", "ORIGIN_AIRPORT": "ORIGIN_AIRPORT_CODE","DESTINATION_AIRPORT": "DESTINATION_AIRPORT_CODE"})
```
